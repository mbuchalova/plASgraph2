```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv("/home/buchalova/school/plASgraph2/results/all_experiemnts_relative_values.csv")
```

```python
table
```

```python
filtered = table[table['f1'] > 0.75]

# Prepare scatter plot
plt.figure()

# Plot each "heads" category separately for legend coloring
for heads_value in sorted(filtered['heads'].unique()):
    subset = filtered[filtered['heads'] == heads_value]
    plt.scatter(subset['precision'], subset['recall'], label=f'heads={heads_value}')

orig_precision = 0.866666666666667
orig_recall = 0.858258928571429
plt.scatter(orig_precision, orig_recall, marker='o')
plt.annotate('original', (orig_precision, orig_recall), textcoords="offset points", xytext=(5,5))

plt.xlabel('Precision')
plt.ylabel('Recall')
plt.legend(title='Number of Heads')

plt.show()
```
