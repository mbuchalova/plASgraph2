```python
first = pd.read_csv("/home/buchalova/school/plASgraph2/results/eval_original_scripts/summary_no_normalization.csv")     # lr = [0.0001,0.0003,0.005]
second = pd.read_csv("/home/buchalova/school/plASgraph2/results/eval_original_scripts/summary_with_normalization.csv")  # lr = [0.0001,0.0003,0.005]
third = pd.read_csv("/home/buchalova/school/plASgraph2/results/eval_original_scripts/summary_combination.csv")          # lr = [0.001,0.003]
```

```python
merged = pd.concat([first,second,third])
```

```python
only_plasmid = merged[merged['mol_type'] == "plasmid"]
```

```python
parts = only_plasmid["method"].str.split("_", expand=True)

heads = parts[0].astype(int)
lr = parts[3].astype(float)
tie = parts[5] == "True"
norm = parts[7]

only_plasmid.insert(2, "heads", heads)
only_plasmid.insert(3, "learning_rate", lr)
only_plasmid.insert(4, "tie", tie)
only_plasmid.insert(5, "normalization", norm)
```
